```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
vendas = pd.read_excel('vendas.xlsx')

vendas.head()
```

```python
le = LabelEncoder()

le.fit(vendas.test)
vendas.test = le.transform(vendas.test)
```

```python
vendas.head(20)
```

```python
x = vendas[['media_rating',	'media_pedidos','test']]
y = vendas['cliente_comprou']
```

```python
from sklearn import tree

arvore_decisao = tree.DecisionTreeClassifier()
```

```python
modelo_ad = arvore_decisao.fit(x, y)
modelo_ad
```

```python
fig = plt.figure(figsize=(70,30))

fig = tree.plot_tree(modelo_ad,
                   feature_names=x.columns,
                   class_names=['Yes', "No"],
                   filled=True)
```
